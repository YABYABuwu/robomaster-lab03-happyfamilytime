# Analyze RoboMaster Logs

สมุดโน้ตบุ๊กสำหรับวิเคราะห์ข้อมูลดิบ (Raw CSV logs) ที่บันทึกมาจาก RoboMaster

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
# ค้นหาไฟล์ CSV ล่าสุดในโฟลเดอร์ data/raw
raw_dir = '../data/raw/run1'
csv_files = glob.glob(os.path.join(raw_dir, '*.csv'))

if csv_files:
    latest_file = max(csv_files, key=os.path.getctime)
    print(f"Loading latest log file: {latest_file}")
    df = pd.read_csv(latest_file)
    display(df.head())
else:
    print("No log files found in data/raw/run1")
```

```python
# โค้ดสำหรับพลอตกราฟวิเคราะห์
if 'df' in locals() and not df.empty:
    plt.figure(figsize=(10, 6))
    # สมมติว่ามีคอลัมน์ชื่อ 'time' และข้อมูล imu
    # plt.plot(df['time'], df['acc_x'], label='Acc X')
    # plt.plot(df['time'], df['acc_y'], label='Acc Y')
    plt.title('RoboMaster IMU Acceleration')
    plt.xlabel('Time (s)')
    plt.ylabel('Acceleration')
    plt.legend()
    plt.grid(True)
    plt.savefig('../analysis/plots/imu_acceleration.png')
    plt.show()
```
